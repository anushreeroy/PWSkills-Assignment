5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility
data_size = 100

# Generating correlated data
x1 = np.random.normal(0, 1, data_size)
x2 = 0.8 * x1 + np.random.normal(0, 0.1, data_size)  # Correlated with x1
x3 = 0.6 * x1 - 0.3 * x2 + np.random.normal(0, 0.2, data_size)  # Correlated with x1 and x2

# Creating a DataFrame
df = pd.DataFrame({'Feature1': x1, 'Feature2': x2, 'Feature3': x3})

# Calculating the correlation matrix
corr_matrix = df.corr()

# Creating the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)

# Adding title
plt.title('Correlation Matrix Heatmap')

# Displaying the heatmap
plt.show()
````

```
x = np.linspace(-5, 5, 100)
 ```

````python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Generating the grid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Creating the DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Creating the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot', autosize=False,
                  width=800, height=800,
                  scene=dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'))

# Displaying the plot
fig.show()
````

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

 counts = [20, 25, 30, 35]

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
import pandas as pd

# Ensure the output displays in the notebook
output_notebook()

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Creating a DataFrame
data = pd.DataFrame({'fruits': fruits, 'counts': counts})

# Creating a ColumnDataSource
source = ColumnDataSource(data)

# Creating the bar chart
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

# Customizing the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Displaying the plot
show(p, notebook_handle=True)

````

5. Create a Bokeh heatmap using the provided dataset.

```
 data_heatmap = np.random.rand(10, 10)
 x = np.linspace(0, 1, 10)
 y = np.linspace(0, 1, 10)
 xx, yy = np.meshgrid(x, y)
 ```

````python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
import numpy as np

# Ensure the output displays in the notebook
output_notebook()

# Generating the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flattening the arrays
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten(),
}

# Setting up the color mapper
mapper = LinearColorMapper(palette='Viridis256', low=data['value'].min(), high=data['value'].max())

# Creating the Bokeh plot
p = figure(title='Heatmap', x_axis_label='X', y_axis_label='Y', width=600, height=600)
p.rect(x='x', y='y', width=1/10, height=1/10, source=data, fill_color=transform('value', mapper), line_color=None)

# Adding color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Displaying the plot
show(p)
````
